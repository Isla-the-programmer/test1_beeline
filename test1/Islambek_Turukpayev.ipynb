```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
#import sklearn.cluster.hierarchical as hclust
from sklearn import preprocessing
import seaborn as sns
import datetime
```

### Импортируем транзакции ken_telecom для визуализации профилей пользователей по месяцам. Необходимо посмотреть изменение данных по времени, для того, чтобы провести сегментный анализ

```python
ken_telecom = pd.read_csv('ken_telecom.csv', sep = ';')
ken_telecom['SESSION_START_TIME'] = ken_telecom['SESSION_START_TIME'].str.split(" ", expand=True,)[0]
```

### Посмотрим численные данные, которые мы можем визуализировать. Добавим колонку ones включающую единицы, чтобы посмотреть количество пользователей  в каждом месяце, пользующихся определенным тарифом. Оплаты за сессию и продолжительность разговора можно представить в виде гистограммы

```python
ken_telecom['ones'] = 1
date_range = ['2019-02-01', '2019-03-01', '2019-04-01', '2019-05-01', '2019-06-01']
print('Распределение тарифов по месяцам')
for i in range(1, len(date_range)):
    plt.figure(figsize=(30, 10), dpi=80)
    tariff = ken_telecom[(ken_telecom['SESSION_START_TIME'] <=date_range[i] ) & (ken_telecom['SESSION_START_TIME'] >= date_range[i-1])][['TARIFF_PLAN_NAME', 'ones']]
    tariff['tariff'] = pd.to_numeric(tariff['TARIFF_PLAN_NAME'].str.split("_", expand =True)[3])
    plt.title(date_range[i])
    plt.xlabel('Тарифные планы')
    plt.ylabel('Количество абонентов')
    sns.barplot(data=tariff.groupby(['tariff']).sum().reset_index(), x = 'tariff', y = 'ones')
    plt.show()
print('Распределение выплат по месяцам')
for i in range(1, len(date_range)):
    plt.figure(figsize=(15, 10), dpi=100)
    pay = ken_telecom[(ken_telecom['SESSION_START_TIME'] <=date_range[i] ) & (ken_telecom['SESSION_START_TIME'] >= date_range[i-1])]['PAY_AMT']
    plt.title(date_range[i])
    plt.xlabel('Сумма списания')
    plt.ylabel('Количество абонентов')
    plt.hist(list(pay), bins = 100)
    plt.show()
print('Распределение продолжительности сессий')
for i in range(1, len(date_range)):
    plt.figure(figsize=(15, 10), dpi=100)
    dur = ken_telecom[(ken_telecom['SESSION_START_TIME'] <=date_range[i] ) & (ken_telecom['SESSION_START_TIME'] >= date_range[i-1])]['SESSION_DURATION_IN_SEC']
    plt.title(date_range[i])
    plt.xlabel('Продолжительность сессий')
    plt.ylabel('Количество абонентов')
    plt.hist(list(dur), bins = 100)
    plt.show()
```

### По распределению видно, что с каждым месяцем падает количество тарифов 64 и растут тарифы 82, 83. При чем абсолютное количество тарифов 82, 83 в 6 месяце практически не падают. Продолжительность сессий и размеры списаний падают в 6 месяце, но это вероятно связано с регистрацией наблюдений в этом месяце

### Рассмотрим корреляции переменных, связанных с пользователями

```python
abon_profile = pd.read_csv('abon_profile.csv', sep =';')
abon_profile = abon_profile.fillna(0)
abon_profile.corr().style.background_gradient(cmap='coolwarm')
```

### Имеются сильные корреляции между некоторыми из переменных. Удалим часть из них, чтобы предотвратить появление кластеров, которые будут иметь большое количество очень близких наблюдений. 

```python
ab1 = abon_profile.iloc[:, ::2]
ab2 = ab1.loc[:,~ab1.columns.isin(['VARIABLE_' + str(i) for i in range(15, 24)])]
ab3 = ab2.loc[:,~ab2.columns.isin(['VARIABLE_' + str(i) for i in range(43, 70)])]
ab3.corr().style.background_gradient(cmap='coolwarm')
```

### Далее, построим кластеры с помощью ближайших соседей, перед этим нормализовав данные

```python
scaler = preprocessing.MinMaxScaler()
features_normal = scaler.fit_transform(ab3.loc[:, ab3.columns != 'ID'])
```

```python
inertia = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(features_normal)
    kmeanModel.fit(features_normal)
    inertia.append(kmeanModel.inertia_)
```

```python
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.show()
```

### Используя 'elbow method', наиболее вероятно подойдут 3 кластера

```python
kmeans = KMeans(n_clusters=3).fit(features_normal)
```

```python
print(len(kmeans.labels_))
```

```python
print(len(abon_profile['ID']))
```

```python
IDs_labels = pd.DataFrame() 
IDs_labels['ID'] = list(abon_profile['ID'])
IDs_labels['labels'] = kmeans.labels_
```

```python
IDs_labels
```

### Построим профили кластеров, рассмотрим дату последней сессии(чтобы посмотреть активность), распределение списаний, распределение тарифных планов

```python
for i in range(3):
    clus = IDs_labels[IDs_labels['labels'] == i].merge(ken_telecom, left_on = 'ID', right_on = 'ID', how = 'inner')
    dates = []
    pay_amt = []
    plt.figure(figsize=(20, 15), dpi=80)
    print(f'{i+1}-Й КЛАСТЕР: ')
    for idd in clus['ID'].unique():
        to_lst = clus[(clus['ID'] == idd) & (clus['SESSION_DURATION_IN_SEC'] != 0) & (clus['DIRECT_IND'] != 1) ]['SESSION_START_TIME'].tolist()
        if len( to_lst ) > 0:
            date = datetime.datetime.strptime(str(max(to_lst)).split(' ')[0], "%Y-%m-%d")
        else: 
            date = '2019-01-01'
        dates.append(date)
    plt.title('Дата последней сессии')
    plt.xlabel('Дата последней сессии')
    plt.ylabel('Количество абонентов')
    plt.hist(dates, bins = 50)
    plt.show()
    print('Распределение выплат по месяцам')
    for i in range(1, len(date_range)):
        plt.figure(figsize=(15, 10), dpi=100)
        pay = clus[(clus['SESSION_START_TIME'] <=date_range[i] ) & (clus['SESSION_START_TIME'] >= date_range[i-1])]['PAY_AMT']
        plt.title(date_range[i])
        plt.xlabel('сумма списания')
        plt.ylabel('Количество абонентов')
        plt.hist(list(pay), bins = 100)
        plt.show()
    clus['ones'] = 1
    print('Распределение тарифов по месяцам')
    for i in range(1, len(date_range)):
        plt.figure(figsize=(30, 10), dpi=100)
        tariff = clus[(clus['SESSION_START_TIME'] <=date_range[i] ) & (clus['SESSION_START_TIME'] >= date_range[i-1])][['TARIFF_PLAN_NAME', 'ones']]
        tariff['tariff'] = pd.to_numeric(tariff['TARIFF_PLAN_NAME'].str.split("_", expand =True)[3])
        plt.title(date_range[i])
        plt.xlabel('Тарифные планы')
        plt.ylabel('Количество абонентов')
        sns.barplot(data=tariff.groupby(['tariff']).sum().reset_index(), x = 'tariff', y = 'ones')
        plt.show()
```

### Рассмотрим для начала третий кластер (номер кластера может меняться с каждой новой кластеризацией, сами кластеры не поменяются), с более тяжелым хвостом последних дат справа. Эти клиенты имеют относительно равное количество списаний с 3 по 6 месяц, также имеют относительно равное распределение тарифов. Этих клиентов можно характеризовать как постоянных. Далее, посмотрим на 1 кластер. Количество списаний и максимальные значения увеличиваются с каждым месяцем, также увеличивается разнообразие тарифов. Причем больше всего изменений произошло в 4 и 5 месяце. Этих клиентов можно характеризовать как новых. Во 2 кластере последние сессии резко падают в 5 и 6 месяце, также падают списания и разнообразие тарифов. Наиболее вероятно, что это уходящие клиенты.
